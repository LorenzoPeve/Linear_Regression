```python
import math
import numpy as np
import pandas as pd

import bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
import scipy.constants
```

```python
phi = scipy.constants.golden
```

```python
n = 100
xi = np.random.uniform(-2,2, size = 100)
xi.shape
```

```python
np.sort(xi)
```

```python
yi = phi*(  (phi**-1 * (xi /(n+1))   / 0.1 )) + 0.1 * np.random.normal(size = n)
yi.shape
```

```python
p = figure()
p.scatter(x = xi, y = yi)
show(p)
```
